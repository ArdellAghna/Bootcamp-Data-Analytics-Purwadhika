## 2. Data Cleaning & Preparation
### Memilih kolom relevan dan membuat dataframe baru

```python
# Memilih kolom utama yang relevan untuk analisis
selected_columns = ['Year', 'quarter', 'city1', 'city2', 'airport_1', 'airport_2', 'passengers', 'fare', 'carrier_lg', 'nsmiles']

# Membuat dataframe baru yang hanya berisikan kolom-kolom di atas
df_clean = df[selected_columns].copy()

# Cek hasilnya
df_clean.head()
```

### Menangani *missing values*, menghapus anomali data, dan memastikan tipe data

```python
# Menghapus baris dengan nilai NaN
df_clean.dropna(inplace=True)

# Menghapus data anomali (Fare <= 0 atau Passengers <= 0)
df_clean = df_clean[(df_clean['fare'] > 0) & (df_clean['passengers'] > 0)]

# Memastikan tipe data sudah benar (Numerik)
df_clean['Year'] = df_clean['Year'].astype(int)
df_clean['quarter'] = df_clean['quarter'].astype(int)
df_clean['passengers'] = df_clean['passengers'].astype(int)
df_clean['fare'] = df_clean['fare'].astype(float)
```

### Menambah Kolom Baru
1. **Route**: Menggabungkan kota dan kode bandara.
2. **Era**: Mengelompokkan tahun berdasarkan peristiwa sejarah: Normal, 9/11, Krisis 2008, Pandemi COVID-19.

```python
# 1. Kolom Route
df_clean['Route'] = df_clean['city1'] + ' (' + df_clean['airport_1'] + ') - ' + df_clean['city2'] + ' (' + df_clean['airport_2'] + ')'

# 2. Membuat Kolom Era
def kategori_era(year):
    if year <= 2000:
        return 'Normal Era (1993-2000)'
    elif 2001 <= year <= 2007:
        return '9/11 Impact & Recovery (2001-2007)'
    elif 2008 <= year <= 2019:
        return 'Financial Crisis & Stability (2008-2019)'
    elif year >= 2020:
        return 'Pandemic & New Normal (2020-2024)'
    return 'Unknown'

df_clean['Era'] = df_clean['Year'].apply(kategori_era)

# Cek hasil penambahan kolom
df_clean[['Year', 'Era', 'Route']].head()
```
