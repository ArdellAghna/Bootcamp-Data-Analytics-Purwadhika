```python
total_umur = int(input("Masukkan jumlah umur Andi dan Budi: "))
rasio_umur = float(input("Rasio umurnya berapa? "))

# rasio = andi/budi
# total = andi + andi * rasio
andi = 1
andi = total_umur/andi * rasio_umur

budi = total_umur - andi

print(f"Umur Andi {round(andi + 2)} dan umur Budi {round(budi + 2)}")
```

```python
jarak_mobil = int(input("Masukkan total jarak mobil A & B: "))
A = int(input("Masukkan kecepatan mobil A: "))
B = int(input("Masukkan kecepatan mobil B: "))
j = int(input("Masukkan jam/start awal: "))

total_kecepatan = A + B
waktu = jarak_mobil / total_kecepatan
jam = int(waktu)
menit = round((waktu - jam) * 60)

print(f"mobil A dan B akan nabrak pada pukul {jam + j}.{menit}")
```

```python
A = int(input("Masukkan jumlah Apel: "))
J = int(input("Masukkan jumlah Jeruk: "))
AN = int(input("Masukkan jumlah Anggur: "))

Apel = A * 10000
Jeruk = J * 15000
Anggur = AN * 20000

Total = Apel + Jeruk + Anggur

print("--------Detail Belanja---------")
print(f"1. Apel   : {A} x 10000 = {Apel}")
print(f"2. Jeruk  : {J} x 10000 = {Jeruk}")
print(f"3. Anggur : {AN} x 10000 = {Anggur}")
print("-------------------------------")
print(f"Total Belanja anda :    {Total} ")
print("-------------------------------")
```

```python
imt = float
hasil = str

m = float(input("Masukkan Massa (kg) :"))
t = float(input("Masukkan Tinggi (cm) :"))
t /= 100
print(f"Massa {m} kg dan tinggi {t} m")

IMT = m / t ** 2
if (IMT < 18.5) :
    hasil = "BERAT BADAN KURANG"
elif (IMT >= 18.5 and IMT <= 24.9) :
    hasil = "BERAT BADAN IDEAL"
elif (IMT >= 25.0 and IMT <= 29.9) :
    hasil = "BERAT BADAN BERLEBIH"
elif (IMT >= 30.0 and IMT <= 39.9) :
    hasil = "BERAT BADAN SANGAT BERLEBIH"
else :
    hasil = "OBESITAS"
    
print(f"IMT = {IMT}, {hasil}!")
```
