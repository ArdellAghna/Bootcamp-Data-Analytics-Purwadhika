```python
stars = ""
n = 5

for i in range(n):
    for j in range(n - i):
        stars += "*"
    stars += "\n"

for i in range(n - 1):
    for j in range((i + 1) + 1):
        stars += "*"
    stars += "\n"

print(stars)
```

```python
stars = ""
n = 5

for i in range(n):
    stars += "  " * (n - i - 1)
    for j in range(2 * i + 1):
        stars += "* "
    stars += "\n"

for i in range(n - 2, -1, -1):
    stars += "  " * (n - i - 1)
    for j in range(2 * i + 1):
        stars += "* "
    stars += "\n"

print(stars)
```

```python
Numbers = [41, 5, 1, 3, 89, 32]

Nmax = Numbers[0]
Nmin = Numbers[0]

j = len(Numbers)

for i in range(j):
    if Nmax <= Numbers[i]:
        Nmax = Numbers[i]
    elif Nmin >= Numbers[i]:
        Nmin = Numbers[i]
print("Maximum Value :", Nmax)
print("Minimum Value :", Nmin)
```

```python
print("----Selamat Datang di Pasar Buah----")

buah = ["Apel", "Jeruk", "Anggur"]
stock = [20, 15, 25]
harga = [10000, 15000, 20000]
buah_keranjang = []
qty = []
harga_keranjang = []

def daftar():
    print("-------------Daftar Buah-----------")
    print("Index   | Nama   | Stock   | Harga")
    for i in range(len(buah)):
        print(f"{i}       | {buah[i]}   | {stock[i]}   | {harga[i]}")

def Cart():
    print("Isi Cart :")
    print("Nama   | Qty   | Harga")
    for i in range(len(buah_keranjang)):
        print(f"{buah_keranjang[i]}   | {qty[i]}   | {harga_keranjang[i]}")

def ucapan_Terima_Kasih_mengunjungi():
    print("-------------------------------")
    print("Terima Kasih telah mengunjungi pasar buah")
    print("-------------------------------")

def ucapan_Terima_Kasih_Sudah_Beli():
    print("-------------------------------")
    print("Terima kasih telah berbelanja buah di toko kami")
    print("-------------------------------")

masukkan = False
while not (masukkan):
    print("List Menu :")
    print(" 1. Menampilkan Daftar Buah")
    print(" 2. Menambah Buah")
    print(" 3. Menghapus Buah")
    print(" 4. Membeli Buah")
    print(" 5. Exit Program")
    a = int(input("Masukkan angka Menu yang ingin dijalankan : "))
    if (a == 1):
        daftar()
        liat_menu = input("Kembali ke menu? (ya/tidak) : ")
        if liat_menu.lower() == "tidak":
            ucapan_Terima_Kasih_mengunjungi()
            break
        
    elif (a == 2):
        B = str(input("Masukkan Nama Buah : "))
        S = int(input("Masukkan Stock Buah : "))
        H = int(input("Masukkan Harga Buah : "))
        buah.append(B)
        stock.append(S)
        harga.append(H)
        print("Buah berhasil ditambahkan.")
        liat_menu = input("Kembali ke menu? (ya/tidak) : ")
        if liat_menu.lower() == "tidak":
            ucapan_Terima_Kasih_mengunjungi()
            break
        
    elif (a == 3):
        daftar()
        hapus = int(input("Masukkan index buah yang ingin dihapus : "))
        if (hapus >= len(buah)):
            print("Index tersebut tidak ada.")
            continue
        buah.pop(hapus)
        stock.pop(hapus)
        harga.pop(hapus)
        print("Buah berhasil dihapus.")
        liat_menu = input("Kembali ke menu? (ya/tidak) : ")
        if liat_menu.lower() == "tidak":
            ucapan_Terima_Kasih_mengunjungi()
            break
    
    elif a == 4:
        while True:
            daftar()
            beli = int(input("Masukkan index buah yang ingin dibeli : "))
            if (beli >= len(buah)):
                print("Index tersebut tidak ada.")
                continue
            jumlah = int(input("Masukkan jumlah yang ingin dibeli : "))
            if (jumlah == 0):
                print("Tidak bisa 0.")
                continue
            elif jumlah > stock[beli]:
                print(f"Stock tidak cukup, stock {buah[beli]} tinggal {stock[beli]}")
            else:
                buah_keranjang.append(buah[beli])
                qty.append(jumlah)
                harga_keranjang.append(harga[beli])
                stock[beli] -= jumlah
                Cart()
            lanjut = input("Mau beli yang lain? (ya/tidak) = ")
            if lanjut.lower() == "tidak":
                break

        print("-------------Daftar Belanja-------------")
        print("Nama   | Qty   | Harga  | Total Harga")
        Total = 0
        for i in range(len(buah_keranjang)):
            total_harga_belanja = qty[i] * harga_keranjang[i]
            Total += total_harga_belanja
            print(f"{buah_keranjang[i]}   | {qty[i]}   | {harga_keranjang[i]}   | {total_harga_belanja}")
        print(f"Total Yang Harus Dibayar   = {Total}")

        if (Total == 0):
            liat_menu = input("Kembali ke menu? (ya/tidak) : ")
            if liat_menu.lower() == "tidak":
                ucapan_Terima_Kasih_mengunjungi()
                break
            else:
                continue
            
        U = int(input("Masukkan jumlah uang anda : "))
        uang = False

        while not(uang):
            if Total == U:
                print("Uang Anda pas")
                ucapan_Terima_Kasih_Sudah_Beli()
                uang = True
        
            elif Total > U:
                sisa = Total - U
                print("-------------------------------")
                print(f"Uang anda kurang sebesar {sisa}")
                U = int(input("Masukkan jumlah uang anda : "))
                if U > Total:
                    s = U - Total
                    print(f"Uang kembalian anda : {s}")
                    ucapan_Terima_Kasih_Sudah_Beli()
                    uang = True
                elif U == Total:
                    print("Uang Anda pas")
                    ucapan_Terima_Kasih_Sudah_Beli()
                    uang = True

            else:
                sisa = U - Total
                print(f"Uang kembalian anda : {sisa}")
                ucapan_Terima_Kasih_Sudah_Beli()
                uang = True

        buah_keranjang.clear()
        qty.clear()
        harga_keranjang.clear()

    elif (a == 5):
        ucapan_Terima_Kasih_mengunjungi()
        break;

    else:
        print("-------------------------------")
        print("Tolong masukan nomer menu yang tersedia!")
        print("-------------------------------")
        continue
```
