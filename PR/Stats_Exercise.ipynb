## **Nomor 3**

* Nilai outlier adalah nilai yang berada di bawah ``Q1 - (1.5 * IQR)`` atau di atas ``Q3 + (1.5 * IQR)``
* Hitung dulu Q1, Q2, dan IQR pada pada kolom **Price** (harga rumah)
* Hitung ada berapa nilai outlier pada kolom **Price**!

```python
import numpy as np

df['Price'] = df['Price'].apply(np.int64)
```

```python
df['Price'] = df['Price'].apply(np.int64)
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

low = Q1 - (1.5 * IQR) 
up = Q3 + (1.5 * IQR)

outL = df[(df['Price'] < low) | (df['Price'] > up)]

len(outL)
```

## **Nomor 4**

#### **``Shapiro Wilk Test``**

Menguji asumsi normalitas variabel ``Price``

#### **Interpretation**
* If the P-Value of the Shapiro Wilk Test is larger than 0.05, we assume a normal distribution
* If the P-Value of the Shapiro Wilk Test is smaller than 0.05, we do not assume a normal distribution

```python
from scipy import stats

# Fungsi mengembalikan (Test Statistic, P-value)
stat, p_value = stats.shapiro(df['Price'])


if p_value > 0.05:
    print(f"P-value: {p_value}. So, We assume a normal distribution.")
    
else:
    print(f"P-value: {p_value}. So, We do not assume a normal distribution.")
```

#### **``D’Agostino and Pearson’s Test``**

Menguji asumsi normalitas variabel ``Price``

#### **Interpretation**
* If the P-Value of the D'Agostino-Pearson Test is larger than 0.05, we assume a normal distribution
* If the P-Value of the D'Agostino-Pearson Test is smaller than 0.05, we do not assume a normal distribution

```python
from scipy import stats

# Fungsi mengembalikan (Test Statistic, P-value)
stat, p_value = stats.normaltest(df['Price'])

if p_value > 0.05:
    print(f"P-value: {p_value}. So, We assume a normal distribution.")
    
else:
    print(f"P-value: {p_value}. So, We do not assume a normal distribution.")
```

## **Nomor 6**

Jika diasumsikan telah memenuhi syarat statistik parametrik, dari beberapa Regionname manakah **Regionname** yang variabel jumlah kamarnya (**Rooms**) memiliki korelasi terkuat dengan variabel harga rumah (**Price**)!

```python
result = []

regions = df['Regionname'].unique()

for region in regions:
    temp = df[df['Regionname'] == region]

    # hitung korelasi Rooms & Price
    corr = temp['Rooms'].corr(temp['Price'])

    # jika korelasi bernilai angka valid (bukan NaN)
    if not np.isnan(corr):
        result.append(region)

df['Regionname'].unique()
```

```python
hasil = {}

for region in regions:
    data_region = df[df['Regionname'] == region]
    hasil[region] = data_region['Rooms'].corr(data_region['Price'])

for regionn,korelasii in hasil.items():
    print(regionn, "=", korelasii)
```

```python
corr_df = (
    pd.DataFrame(list(hasil.items()), columns=['', 'Correlation Price vs Rooms'])
    .sort_values(by='Correlation Price vs Rooms', ascending=False)
    .reset_index(drop=True) # untuk memulai dari 0
)

corr_df
```

## **Nomor 8**

Jika diasumsikan telah memenuhi syarat statistik parametrik, dari beberapa Regionname manakah **Regionname** yang variabel jumlah jarak ke CBD (**Distance**) paling tidak berkorelasi dengan variabel harga rumah (**Price**)! ## **Nomor 9**

Uji hipotesis apakah metode pembelian/**Method** berkaitan dengan **Type** / tipe rumah (dependen) atau tidak saling berkaitan (independen)!

```python
table = pd.crosstab(df['Type'], df['Method'])
table
```

#### **Hypothesis:**
* Ho : The two variables are independent
* Ha :  The two variables are dependent.

```python
# uji coba untuk hipotesis yang datanya kategorikal
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table)

print(f"P value: {p}")

if p < 0.05:
    print("Dependent (tolak H0)")
    
else:
    print("Independent (terima H0)")
```

## **Nomor 10**

Uji hipotesis apakah rata-rata harga rumah (**Price**) antar Regionname berbeda ataukah sama!

Diasumsikan persyaratan statistik parametrik terpenuhi

**Hipotesis**:
* Ho : μ1 = μ2 = … = μk (Rata-rata harga rumah antar Regionname sama)  
* Ha : Setidaknya ada satu pasang Regionname yang rata-rata harga rumahnya tidak sama, tapi kita tidak tahu yang mana

```python
from scipy import stats

# 1. Cek semua Regionname unik
regions = df['Regionname'].unique()
print(regions)
```

```python
price_groups = [df[df['Regionname'] == region]['Price'] for region in regions]

stats.f_oneway(*price_groups)

# anova untuk mencari perbandingan rata" antara beberapa kelompok (dalam konteks ini price antar region)
```

Jika p-value kurang dari sama dengan 0.05 maka cukup bukti untuk menolak Ho.
Berarti Ha diterima ada pasangan Regionname yang perbedaan rata-rata harga rumahnya signifikan, tapi tidak tahu yang mana.

```python
price_mean = (
    df.groupby('Regionname')['Price'].mean()
    .reset_index()
    .rename(columns={'Price': 'Price Mean'})
    .sort_values(by='Price Mean', ascending=False)
    .reset_index(drop=True)
)
price_mean
```
